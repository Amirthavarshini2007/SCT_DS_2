```python
from google.colab import files
uploaded = files.upload()
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style="whitegrid")
data = pd.read_csv("WineQT.csv", sep=',')
data.head()
data.info()
data.describe()
data.isnull().sum()
data = data.drop_duplicates()
plt.figure(figsize=(6,4))
sns.countplot(x='quality', data=data)
plt.title("Distribution of Wine Quality")
plt.show()
features = ['alcohol', 'volatile acidity', 'citric acid', 'residual sugar']
for feature in features:
    plt.figure(figsize=(8,5))
    sns.boxplot(x='quality', y=feature, data=data)
    plt.title(f"{feature} by Wine Quality")
    plt.show()
sns.pairplot(
    data,
    vars=['alcohol', 'volatile acidity', 'citric acid', 'residual sugar'],
    hue='quality',
    diag_kind='kde'
)
plt.show()
plt.figure(figsize=(12,8))
corr_matrix = data.corr()
sns.heatmap(
    corr_matrix,
    annot=True,
    cmap="coolwarm",
    fmt=".2f"
)
plt.title("Correlation Heatmap")
plt.show()
for feature in features:
    plt.figure(figsize=(6,4))
    sns.histplot(data[feature], kde=True, bins=20)
    plt.title(f"Distribution of {feature}")
    plt.show()
```
